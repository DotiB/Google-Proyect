I import scipy to search for possible outliers

```python
from scipy import stats

# Select numeric columns (float64 and int64)
numeric_columns = data_filtrado.select_dtypes(include=['float64', 'int64']).columns

# Create a DataFrame to store the outliers
outliers = pd.DataFrame()

# Iterate over the numeric columns
for column in numeric_columns:
    # Calculate the z-scores for the current column
    z_scores = stats.zscore(data_filtrado[column])
    
    # Set the threshold to identify outliers (3 standard deviations)
    threshold = 3
    
    # Select values that are outliers based on the z-score
    outliers[column] = data_filtrado[column][(z_scores > threshold) | (z_scores < -threshold)]

# Print the DataFrame of outliers
print("Outliers:")
print(outliers)
```

I search for outliers with the 'Income' variable

```python
# Calculate the mean and standard deviation of the 'Income' column
mean_income = data_filtrado['Income'].mean()
std_income = data_filtrado['Income'].std()

# Set a threshold to identify outliers (e.g., 4 standard deviations)
upper_threshold_income = mean_income + 4 * std_income
lower_threshold_income = mean_income - 4 * std_income

# Identify outliers in the 'Income' column
outliers_income = data_filtrado[(data_filtrado['Income'] > upper_threshold_income) | (data_filtrado['Income'] < lower_threshold_income)]

# Print the 'Income' values for the identified outliers
print("Income Outliers:")
print(outliers_income[['Income']])

# Configure the figure size for the scatter plot
plt.figure(figsize=(10, 6))

# Create a scatter plot with all data points (in blue)
plt.scatter(data_filtrado['Income'], data_filtrado['MntWines'], alpha=0.3, color='blue', label='Data')

# Highlight outliers in red
plt.scatter(outliers_income['Income'], outliers_income['MntWines'], color='red', label='Outliers')

# Set the title and axis labels
plt.title('Scatter Plot of Income with Outliers')
plt.xlabel('Income')
plt.ylabel('MntWines')

# Show the legend and activate the grid
plt.legend()
plt.grid(True)

# Show the scatter plot
plt.show()

```

I repeat the process with the 'Year_Birth' variable

```python
# Calculate the mean and standard deviation of the 'Year_Birth' column
mean_year_birth = data_filtrado['Year_Birth'].mean()
std_year_birth = data_filtrado['Year_Birth'].std()

# Set a threshold to identify outliers (e.g., 4 standard deviations)
upper_threshold_year_birth = mean_year_birth + 4 * std_year_birth
lower_threshold_year_birth = mean_year_birth - 4 * std_year_birth

# Identify outliers in the 'Year_Birth' column
outliers_year_birth = data_filtrado[
    (data_filtrado['Year_Birth'] > upper_threshold_year_birth) | (data_filtrado['Year_Birth'] < lower_threshold_year_birth)]

# Print the 'Year_Birth' values for the identified outliers
print("Year_Birth Outliers:")
print(outliers_year_birth[['Year_Birth']])

# Configure the figure size for the scatter plot
plt.figure(figsize=(10, 6))

# Create a scatter plot with all data points (in blue)
plt.scatter(data_filtrado['Year_Birth'], data_filtrado['MntWines'], alpha=0.3, color='blue', label='Data')

# Highlight outliers in red
plt.scatter(outliers_year_birth['Year_Birth'], outliers_year_birth['MntWines'], color='red', label='Outliers')

# Set the title and axis labels
plt.title('Scatter Plot of Year_Birth with Outliers')
plt.xlabel('Year_Birth')
plt.ylabel('MntWines')

# Show the legend and activate the grid
plt.legend()
plt.grid(True)

# Show the scatter plot
plt.show()
```

```python
# Merge the data frames 'data_filtrado' and 'outliers' using a left join
merged_df = pd.merge(data_filtrado, outliers, how='left', indicator=True)

# Filter rows where the indicator column is 'left_only' (rows only in 'data_filtrado')
data_sin_outliers = merged_df[merged_df['_merge'] == 'left_only']

# Drop the indicator column as it's no longer needed
data_sin_outliers = data_sin_outliers.drop(columns=['_merge'])

# Display the resulting DataFrame without outliers
data_sin_outliers
```
